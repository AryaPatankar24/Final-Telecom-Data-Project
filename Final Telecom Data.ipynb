```python
service_features = ['PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies']
```

```python
def churn_by_services(df):
  service_features = ['PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies']

  for i in service_features:
    plt.figure(figsize = (6,4))

    #Try without hue parameter
    sns.countplot(data = df, x = i,hue = 'Churn', palette = 'viridis')
    plt.title(f"Churn Raye by {i}")  #corrected title formaating
    plt.show()
```

```python
churn_by_services(df)
```

```python
def count_by_services(df):
  service_features = ['PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies']

  for i in service_features:
    churn_count = df.groupby(i)['Churn'].value_counts().unstack()
    print(f"Column {i} value count :- {churn_count}")
```

```python
count_by_services(df)
```

#27. Can you create new features that might be predictive of churn? For example:
o The total number of services a customer has.


o The average charge per service.


o The number of months a customer has been on their current contract.


o Binned tenure groups (e.g., 0-12 months, 13-24 months, etc.).



o Categories for monthly charges (e.g., low, medium, high).

```python
df['Tenure_group'].unique()
```

```python
df.head(1)
```

```python
df['Num_Services'].unique()
```

```python
df['Avg_Charge_Per_Service']= df['MonthlyCharges'] / df ['Num_Services']
```

```python
df['MonthlyCharges'] = df['tenure']
```

```python
def bin_tenure(tenure):
  if tenure <= 12:
    return '0-12 months'
  elif tenure <= 24:
    return '13=24 months'
  elif tenure <= 36:
    return '25=36 months'
  elif tenure <= 48:
    return '37-48 months'
  else:
    return '49+ months'

  df['TenureBin'] = df ['tenure'].apply(bin_tenure)
```

```python
df['MonthlyCharges'].describe()
```

```python
def category_charges(i):
  if i <= df['MonthlyCharges'].quantile(0.33):
    return 'Low'
  elif i<= df ['MonthlyCharges'].quantile(0.66):
    return 'Medium'
  else:
    return 'High'

  df['MonthlyCharges_Category'] = df['MonthlyCharges'].apply(category_charges)


```
